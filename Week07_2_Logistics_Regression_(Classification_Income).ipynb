```python
# Function to calculate percentage of 0s in each column
def calculate_zero_percentage(df):
    percentages = {}
    for col in df.columns:
        # Count the number of 0s in the column
        zero_count = (df[col] == 0).sum()
        # Calculate percentage
        percentage = (zero_count / len(df)) * 100
        percentages[col] = percentage
    return percentages

# Get the percentages for all columns
zero_percentages = calculate_zero_percentage(df)

# Display the result
for col, percentage in zero_percentages.items():
    print(f"Percentage of 0s in column '{col}': {percentage:.2f}%")
```

# Rewrite all function

```python
def build_and_evaluate_logistic_regression(df, target_column, feature_columns):
  """
  Builds and evaluates a logistic regression model.

  Args:
    df: The pandas DataFrame containing the data.
    target_column: The name of the column representing the target variable.
    feature_columns: A list of column names to be used as features.
  """
  x = df[feature_columns]
  y = df[target_column]

  x_normalize = preprocessing.StandardScaler()
  x_norm = x_normalize.fit_transform(x)

  x_train, x_test, y_train, y_test = model_selection.train_test_split(
      x_norm, y, test_size=0.3, random_state=101, stratify=y)

  log_model = linear_model.LogisticRegression()
  log_model.fit(x_train, y_train)

  # Get the coefficients and intercept of the model
  coefficients = log_model.coef_
  intercept = log_model.intercept_

  # Store the equation
  equation_str = "log(odds) = " + str(round(intercept[0], 4))
  for i in range(len(coefficients[0])):
    equation_str += " + " + str(round(coefficients[0][i], 4)) + " * x" + str(i)

  # Store the variable mapping
  variable_mapping = {}
  for i in range(len(coefficients[0])):
    variable_mapping[f"x{i}"] = x.columns[i]

  y1_pred = log_model.predict(x_test)

  # Count the occurrences of each unique value in y1_pred
  counts = Counter(y1_pred)

  # Generate the classification report
  report = classification_report(y_test, y1_pred, digits=4)

  return equation_str, variable_mapping, report


# Example usage:
target_column = 'income'
feature_columns = ['age', 'workclass', 'education', 'occupation']
equation, variable_mapping, report = build_and_evaluate_logistic_regression(
    df, target_column, feature_columns)


# Print the equation, variable mapping, and report
print("Equation:", equation)
print("\nVariable Mapping:", variable_mapping)
print("\nClassification Report:\n", report)
```
